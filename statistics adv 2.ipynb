12. What is t test and how it is used in hypothesis testing?
 
   T-test is a statistical hypothesis test used to compare the means of two groups to determine if there's a statistically significant difference between them.
How T-tests are used in Hypothesis Testing:
1. Formulate Hypotheses:
Null Hypothesis (H0): Assumes there is no significant difference between the means of the two groups. 
Alternative Hypothesis (H1): Assumes there is a significant difference between the means of the two groups. 
2. Choose the appropriate t-test:
One-sample t-test: Used to compare a single sample mean to a known population mean. 
Independent samples t-test: Used to compare the means of two independent groups. 
Paired samples t-test: Used to compare the means of two related groups, such as "before and after" measurements on the same subjects. 
3. Calculate the t-statistic:
A formula is used to calculate a t-statistic based on the sample data. 
4. Determine the p-value:
The p-value represents the probability of observing a t-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true. 
5. Make a decision:
If the p-value is less than the chosen significance level (alpha, typically 0.05), the null hypothesis is rejected, suggesting a statistically significant difference between the groups. 
If the p-value is greater than or equal to the significance level, the null hypothesis is not rejected.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete uniform distribution
start = 1   # Minimum value (e.g., 1 for a die)
end = 6     # Maximum value (e.g., 6 for a die)

# Create the values in the distribution
values = np.arange(start, end + 1)

# Since it's a uniform distribution, each value has equal probability
probability = 1 / len(values)

# Calculate the CDF
cdf = np.cumsum([probability] * len(values))

# Plot the CDF
plt.step(values, cdf, where='post', color='green', label='CDF')

# Add labels and title
plt.title('Cumulative Distribution Function (Discrete Uniform)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.ylim(0, 1.1)
plt.xticks(values)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Generate example data (normal distribution)
np.random.seed(0)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Calculate z-scores
z_scores = calculate_z_scores(data)

# Plot z-scores as a histogram
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

14. Write a python program to calculate the z score for a given data point ans compare it to standard normal distribution.

```python
import numpy as np
from scipy.stats import norm

def compute_z_score(x, data):
    """
    x: specific data point
    data: list or array of values
    Returns: (z_score, p_lower, p_upper, pdf_value)
    """
    mu = np.mean(data)
    sigma = np.std(data, ddof=0)  # population std
    z = (x - mu) / sigma

    p_lower = norm.cdf(z)        # P(Z  z)
    p_upper = 1 - p_lower        # P(Z > z)
    pdf_val = norm.pdf(z)        # density at z

    return z, p_lower, p_upper, pdf_val


# Example usage:
data = [70, 75, 80, 85, 90, 95, 100]
x = 85

z, p_lower, p_upper, pdf_val = compute_z_score(x, data)

print(f"Data: {data}")
print(f"Point x = {x}")
print(f"Mean = {np.mean(data):.2f}, Std = {np.std(data, ddof=0):.2f}")
print(f"Z-score = {z:.3f}")
print(f"P(Z  {z:.3f}) = {p_lower:.3f}")
print(f"P(Z > {z:.3f}) = {p_upper:.3f}")
print(f"PDF at Z = {z:.3f}: {pdf_val:.3f}")
```

15. Implement hyphothesis testing using z statistics for a sample data set.

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test for the mean of one sample.
    Returns: z_stat, p_value, conclusion
    """
    n = len(data)
    sample_mean = np.mean(data)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed test

    conclusion = ("Reject H0" if p_value < alpha
                  else "Fail to reject H0")
    return z_stat, p_value, conclusion

# Example usage:
data = [11.8] * 100           # e.g., sample battery life
pop_mean = 12.0              # claimed average
pop_std = 0.5                # known population std dev

z, p, result = one_sample_z_test(data, pop_mean, pop_std)
print(f"Z = {z:.3f}, p-value = {p:.3f}  {result}")
```

16. Create a confidence interval for a dataset using python and interpret the result.

```python
import numpy as np
from scipy import stats

def compute_confidence_interval(data, confidence=0.95):
    """
    Returns: (mean, lower_bound, upper_bound)
    Uses t-distribution if n < 30, else uses normal distribution.
    """
    arr = np.array(data)
    n = len(arr)
    mean = np.mean(arr)
    sem = stats.sem(arr)  # standard error = s / sqrt(n)

    if n < 30:
        # small sample: use t-distribution
        crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)
        method = "t-distribution"
    else:
        # large sample: use normal distribution
        crit = stats.norm.ppf((1 + confidence) / 2)
        method = "z-distribution"

    margin = crit * sem
    return mean, mean - margin, mean + margin, method

# Example usage:
data = [10, 12, 14, 11, 15, 13, 12, 14, 10, 13]  # sample of size n=10
mean, lower, upper, method = compute_confidence_interval(data, confidence=0.95)

print(f"Sample size: {len(data)}  using {method}")
print(f"Mean = {mean:.2f}")
print(f"95% CI: [{lower:.2f}, {upper:.2f}]")
```

```python
22. Write a python function to calculate the skewness and kurtosis of a dataset generated from a normal distrubution.
```

```python
import numpy as np
from scipy.stats import skew, kurtosis, norm
import matplotlib.pyplot as plt

def analyze_distribution(size=1000, mu=0, sigma=1):
    # 1. Simulate normal data
    data = np.random.normal(loc=mu, scale=sigma, size=size)
    
    # 2. Calculate skewness and kurtosis
    sk = skew(data, bias=False)                       # sample skewness
    kt = kurtosis(data, fisher=True, bias=False)      # excess kurtosis (normal = 0)
    
    # 3. Show results
    print(f"Sample size: {size}")
    print(f"Simulated from N({mu}, {sigma})")
    print(f"Skewness: {sk:.3f}")
    print(f"Excess Kurtosis: {kt:.3f}")
    print()
    
    # 4. Interpret
    if abs(sk) < 0.1:
        print(" Skewness close to 0: distribution is roughly symmetric.")
    else:
        print(" Skewness noticeably deviates from 0: potential asymmetry.")
    
    if abs(kt) < 0.5:
        print(" Excess kurtosis near 0: tail weight is similar to a normal distribution.")
    elif kt > 0:
        print(" Positive excess kurtosis: heavy tails (leptokurtic).")
    else:
        print(" Negative excess kurtosis: light tails (platykurtic).")
    
    # 5. Optional plot
    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='k')
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
    plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Normal PDF')
    plt.title("Histogram with Theoretical Normal PDF")
    plt.legend()
    plt.show()

# Example usage
analyze_distribution(size=1000, mu=0, sigma=1)
```

1.Write a python program to perform a z test for compairing a sample mean to a known population mean and imterpret the results.

```python
# Stepbystep Z-test in Python  beginner version

import numpy as np
from statsmodels.stats.weightstats import ztest

def perform_z_test(sample, population_mean, alpha=0.05, alternative='two-sided'):
    """
    sample           : list or array of sample observations
    population_mean : known mean of the population ()
    alpha            : significance level (default 5%)
    alternative      : 'two-sided', 'larger', or 'smaller'
    """
    # Convert sample to numpy array
    data = np.array(sample)
    n = len(data)
    sample_mean = data.mean()
    
    # Compute Z-statistic and p-value using statsmodels
    z_stat, p_value = ztest(data, value=population_mean, alternative=alternative)
    
    # Print summary
    print(f"Sample size       : {n}")
    print(f"Sample mean       : {sample_mean:.4f}")
    print(f"Population mean (): {population_mean}")
    print(f"Z-statistic       : {z_stat:.4f}")
    print(f"P-value           : {p_value:.4f}")
    
    # Interpret results
    if p_value < alpha:
        print(f" Reject H at ={alpha}: sample mean is significantly different.")
    else:
        print(f" Fail to reject H at ={alpha}: no significant difference.")

# ===== Example usage =====
if __name__ == "__main__":
    # Example: known population mean = 100
    sample_data = [102.45, 101.48, 99.28, 100.78, 102.74,
                   97.59, 97.19, 101.83, 98.58, 98.46]
    perform_z_test(sample_data, population_mean=100)
```

2. Stimulate a random data to perform hyphotheis testing and calculate the corresponding p value using python.

```python
import numpy as np
from scipy import stats

def simulate_and_test(population_mean=50, sample_size=30, alpha=0.05):
    """
    Simulates a random sample from a normal distribution,
    performs a t-test against a known population mean,
    and interprets the p-value.
    
    Args:
        population_mean (float): the hypothesized true mean ().
        sample_size (int): number of observations to simulate.
        alpha (float): significance level (e.g., 0.05).
    """
    # 1. Generate random sample data (mean around population_mean, sd=10)
    sample = np.random.normal(loc=population_mean, scale=10, size=sample_size)
    
    # 2. Perform one-sample t-test
    t_stat, p_value = stats.ttest_1samp(sample, population_mean)
    
    # 3. Report results
    print(f"Sample size       : {sample_size}")
    print(f"Sample mean       : {sample.mean():.3f}")
    print(f"T-statistic       : {t_stat:.3f}")
    print(f"P-value           : {p_value:.4f}")
    
    # 4. Interpret p-value
    if p_value < alpha:
        print(f" p-value < {alpha}: Reject H  sample mean differs significantly from {population_mean}.")
    else:
        print(f" p-value  {alpha}: Fail to reject H  no significant evidence that sample mean differs from {population_mean}.")

if __name__ == "__main__":
    simulate_and_test()
```

3. Implement a one sample z test using python to compare the sample mean with population.

```python
import numpy as np
from scipy import stats
import math

def one_sample_z_test(sample, pop_mean, pop_sigma, alpha=0.05):
    """
    sample     : list or NumPy array of observations
    pop_mean   : known population mean ()
    pop_sigma  : known population standard deviation ()
    alpha      : significance level (default = 0.05)
    """
    n = len(sample)
    x_bar = np.mean(sample)
    se = pop_sigma / math.sqrt(n)
    z = (x_bar - pop_mean) / se                  # Zstatistic
    p = 2 * (1 - stats.norm.cdf(abs(z)))        # two-sided pvalue

    print(f"n = {n}, sample mean = {x_bar:.3f}")
    print(f"Z = {z:.3f}, pvalue = {p:.4f}")
    if p < alpha:
        print(f" p < {alpha}: Reject H  mean  {pop_mean}")
    else:
        print(f" p  {alpha}: Fail to reject H  no evidence mean  {pop_mean}")

# Example usage:
if __name__ == "__main__":
    # Simulate data: true mean is 100, true sigma is 15
    np.random.seed(1)
    data = np.random.normal(loc=100, scale=15, size=50)
    one_sample_z_test(data, pop_mean=100, pop_sigma=15)
```

4. Perform a two tailed z test using python and visualize decision region on a plot.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def two_tailed_z_test(sample, pop_mean, pop_sigma, alpha=0.05):
    n = len(sample)
    xbar = np.mean(sample)
    se = pop_sigma / math.sqrt(n)
    z = (xbar - pop_mean) / se
    p = 2 * (1 - norm.cdf(abs(z)))

    print(f"n = {n}, sample mean = {xbar:.3f}")
    print(f"Z-statistic = {z:.3f}, p-value = {p:.4f}")
    print("Decision:", "Reject H" if p < alpha else "Fail to reject H")

    # Plot
    x = np.linspace(-4, 4, 400)
    y = norm.pdf(x)
    z_crit = norm.ppf(1 - alpha/2)

    plt.plot(x, y, label='Standard Normal')
    plt.fill_between(x, y, where=(x <= -z_crit)|(x >= z_crit), color='red', alpha=0.3)
    plt.axvline(-z_crit, color='k', linestyle='--', label=f'{z_crit:.2f} critical')
    plt.axvline(z_crit, color='k', linestyle='--')
    plt.axvline(z, color='blue', linewidth=2, label=f'z = {z:.2f}')
    plt.title('Two-Tailed Z-Test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    np.random.seed(1)
    data = np.random.normal(loc=100, scale=15, size=30)  # simulated sample
    two_tailed_z_test(data, pop_mean=100, pop_sigma=15)
```

5. Create a python function that calculate and visualize type 1 and type 2 error during hyphothesisb testing. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def plot_type1_type2(pop_mean, alt_mean, pop_sigma, n, alpha=0.05):
    """
    Plots Type I and Type II error regions for a two-tailed Z-test.
    
    Args:
        pop_mean (float): mean under H0
        alt_mean (float): mean under Ha
        pop_sigma (float): known population std dev
        n (int): sample size
        alpha (float): significance level
    """
    se = pop_sigma / math.sqrt(n)
    z_crit = norm.ppf(1 - alpha/2)
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 500)
    
    # Null and alternative PDFs
    pdf0 = norm.pdf(x, loc=pop_mean, scale=se)
    pdf1 = norm.pdf(x, loc=alt_mean, scale=se)
    
    # Critical boundaries
    lower = pop_mean - z_crit * se
    upper = pop_mean + z_crit * se
    
    # Compute error probabilities
    alpha_area = norm.sf(upper, loc=pop_mean, scale=se) + norm.cdf(lower, loc=pop_mean, scale=se)
    beta_area = norm.cdf(upper, loc=alt_mean, scale=se) - norm.cdf(lower, loc=alt_mean, scale=se)
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf0, label="H ()")
    plt.plot(x, pdf1, label="H ()")
    
    plt.fill_between(x, pdf0, where=(x <= lower)|(x >= upper), color="red", alpha=0.3, label=f"Type I ({alpha_area:.2f})")
    plt.fill_between(x, pdf1, where=(x >= lower)&(x <= upper), color="magenta", alpha=0.3, label=f"Type II ({beta_area:.2f})")
    
    plt.axvline(lower, linestyle='--', color='black', label=f"Critical = {lower:.2f}, {upper:.2f}")
    plt.axvline(upper, linestyle='--', color='black')
    plt.title("Type I & Type II Error Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
    
    print(f"Calculated   {alpha_area:.3f}")
    print(f"Calculated   {beta_area:.3f}")
    print(f"Test Power = 1 -   {1 - beta_area:.3f}")

# Example usage:
if __name__ == "__main__":
    np.random.seed(0)
    plot_type1_type2(
        pop_mean=100,     # H mean
        alt_mean=105,     # true mean under H
        pop_sigma=10,     # known 
        n=30,             # sample size
        alpha=0.05        # significance level
    )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel
import seaborn as sns

def paired_t_test_and_plot(before, after, alpha=0.05):
    """
    Performs paired t-test and visualizes the comparison.

    Args:
        before (array-like): paired sample before treatment
        after (array-like): paired sample after treatment
        alpha (float): significance level (default 0.05)
    """
    before = np.array(before)
    after = np.array(after)
    diff = after - before

    # Paired t-test
    t_stat, p_value = ttest_rel(before, after)
    print(f"Paired t-statistic = {t_stat:.3f}")
    print(f"P-value (two-tailed) = {p_value:.4f}")
    print("", "Reject H: means differ" if p_value < alpha else "Fail to reject H: no difference")

    # Visualization
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    # Scatter plot: before vs after
    axes[0].scatter(before, after, color="blue")
    max_val = max(np.max(before), np.max(after))
    axes[0].plot([0, max_val], [0, max_val], color="gray", linestyle="--")
    axes[0].set_title("Before vs After")
    axes[0].set_xlabel("Before")
    axes[0].set_ylabel("After")

    # Distribution of differences
    sns.boxplot(y=diff, ax=axes[1], color="lightgreen")
    axes[1].axhline(0, color="red", linestyle="--")
    axes[1].set_title("Differences (After - Before)")
    axes[1].set_ylabel("Difference")

    plt.suptitle(f"Paired t-test (t = {t_stat:.2f}, p = {p_value:.4f})")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    before = np.random.normal(50, 5, size=30)
    after = before + np.random.normal(3, 5, size=30)  # simulate an increase
    paired_t_test_and_plot(before, after)
```

7. Stimulate data and perform both z test and t test then compare the result using python .

```python
import numpy as np
import scipy.stats as stats
import math

# Simulate sample data
np.random.seed(0)
sample_size = 30
population_mean = 100
population_std = 15
sample_data = np.random.normal(population_mean, population_std, sample_size)

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
standard_error = sample_std / math.sqrt(sample_size)

# Z-test (population std known)
z_stat = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
p_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed p-value

# T-test (population std unknown)
t_stat, p_t = stats.ttest_1samp(sample_data, population_mean)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-test: Z = {z_stat:.2f}, p-value = {p_z:.4f}")
print(f"T-test: t = {t_stat:.2f}, p-value = {p_t:.4f}")

# Conclusion
alpha = 0.05
if p_z < alpha:
    print("Z-test: Reject null hypothesis (significant difference)")
else:
    print("Z-test: Fail to reject null hypothesis (no significant difference)")

if p_t < alpha:
    print("T-test: Reject null hypothesis (significant difference)")
else:
    print("T-test: Fail to reject null hypothesis (no significant difference)")
```

8. Write a python function to calculate confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
        data (array-like): Sample data
        confidence_level (float): Confidence level (default is 0.95 for 95%)

    Returns:
        tuple: Lower and upper bounds of the confidence interval
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(data)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Critical value from the t-distribution
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Margin of error
    margin_of_error = critical_value * standard_error

    # Confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval
```

```python
# Sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)

# Calculate 95% confidence interval
ci = calculate_confidence_interval(sample_data, confidence_level=0.95)
print(f"95% Confidence Interval: {ci}")
```

9. Implement a Bayesian inference method using Bayes theorem in python and explain the process. 

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): Prior probability P(H)
        likelihood (float): Likelihood P(E|H)
        false_positive_rate (float): False positive rate P(E|H)

    Returns:
        float: Posterior probability P(H|E)
    """
    # Probability of evidence (E)
    p_e = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Posterior probability
    p_h_given_e = (likelihood * prior) / p_e
    return p_h_given_e

# Given values
prior = 0.01  # P(H)
likelihood = 0.95  # P(E|H)
false_positive_rate = 0.05  # P(E|H)

# Calculate posterior probability
posterior = bayes_theorem(prior, likelihood, false_positive_rate)
print(f"Posterior Probability: {posterior:.4f}")
```

10. Perform a chi square test for independence betweeen two categorical variables in python.

| Gender | Smokes | Doesn't Smoke |
| ------ | ------ | ------------- |
| Male   | 30     | 70            |

| Female | 20     | 80            |


```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed data (contingency table)
observed = np.array([[30, 70], [20, 80]])

# Perform Chi-Square Test
chi2_stat, p_val, dof, expected = chi2_contingency(observed)

# Display results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpret the result
alpha = 0.05
if p_val <= alpha:
    print("Reject H: There is a significant association between gender and smoking habit.")
else:
    print("Fail to reject H: There is no significant association between gender and smoking habit.")
```

11. Write a python program to calculate the expected frequencies for a chi square test based on observed data.

```python
import numpy as np

# Observed frequencies
observed = np.array([[30, 10], [20, 40]])

# Calculate row totals, column totals, and grand total
row_totals = observed.sum(axis=1, keepdims=True)
column_totals = observed.sum(axis=0, keepdims=True)
grand_total = observed.sum()

# Calculate expected frequencies
expected = (row_totals @ column_totals) / grand_total

# Display the expected frequencies
print("Expected Frequencies:")
print(expected)
```

12.Perform a goodness of fit test using python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([8, 10, 12, 11, 9, 10])

# Expected frequencies (assuming a fair die)
expected = np.array([10, 10, 10, 10, 10, 10])

# Perform Chi-Square Goodness of Fit Test
chi2_stat, p_val = chisquare(observed, expected)

# Display results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val <= alpha:
    print("Reject H: The observed frequencies significantly differ from the expected frequencies.")
else:
    print("Fail to reject H: The observed frequencies do not significantly differ from the expected frequencies.")
```
